1. Create a variable phrase containing a list of words. Experiment with the operations described in this chapter, including addition, multiplication, indexing, slicing, and sorting.

```python
# Create a variable 'phrase' containing a list of words
phrase = ['apple', 'banana', 'cherry', 'date', 'fig', 'grape']

# Addition: Concatenate two lists
more_fruits = ['honeydew', 'kiwi', 'lemon']
combined_fruits = phrase + more_fruits
print("Combined Fruits:", combined_fruits)

# Multiplication: Repeat the list
repeated_fruits = phrase * 2
print("Repeated Fruits:", repeated_fruits)

# Indexing: Access a specific element
first_fruit = phrase[0]
print("First Fruit:", first_fruit)

# Negative Indexing: Access an element from the end
last_fruit = phrase[-1]
print("Last Fruit:", last_fruit)

# Slicing: Get a sublist of the first 3 elements
first_three_fruits = phrase[:3]
print("First Three Fruits:", first_three_fruits)

# Slicing with a step: Get every second element
every_second_fruit = phrase[::2]
print("Every Second Fruit:", every_second_fruit)

# Sorting: Sort the list alphabetically
sorted_phrase = sorted(phrase)
print("Sorted Fruits:", sorted_phrase)

# Reverse sorting: Sort the list in reverse alphabetical order
sorted_phrase_reverse = sorted(phrase, reverse=True)
print("Sorted Fruits (Reverse):", sorted_phrase_reverse)

# Sorting in place: Modify the original list to be sorted
phrase.sort()
print("Phrase Sorted In-Place:", phrase)

# Reverse in place: Modify the original list to be reversed
phrase.reverse()
print("Phrase Reversed In-Place:", phrase)
```

2.Use the corpus module to explore austen-persuasion.txt. How many word tokens does this book have? How many word types?

```python
import nltk
from nltk.corpus import gutenberg

# Make sure the necessary NLTK data is downloaded
nltk.download('gutenberg')

# Load the text from the Gutenberg corpus
text = gutenberg.raw('austen-persuasion.txt')

# Tokenize the text
tokens = nltk.word_tokenize(text)

# Count the number of tokens
num_tokens = len(tokens)

# Get the set of unique tokens (word types)
types = set(tokens)
num_types = len(types)

print(f"Number of word tokens: {num_tokens}")
print(f"Number of word types: {num_types}")
```

3. Use the Brown Corpus reader nltk.corpus.brown.words() or the Web Text Corpus reader nltk.corpus.webtext.words() to access some sample text in two different genres.

```python
#Using the Brown Corpus

import nltk
from nltk.corpus import brown

# Make sure the necessary NLTK data is downloaded
nltk.download('brown')

# Access words from two different genres
news_words = brown.words(categories='news')
fiction_words = brown.words(categories='fiction')

# Convert lists to sets to get unique words
news_types = set(news_words)
fiction_types = set(fiction_words)

# Print the number of words and types in each genre
print(f"Number of words in news genre: {len(news_words)}")
print(f"Number of unique words (types) in news genre: {len(news_types)}")
print(f"Number of words in fiction genre: {len(fiction_words)}")
print(f"Number of unique words (types) in fiction genre: {len(fiction_types)}")
```

```python
#Using the Web Text Corpus

import nltk
from nltk.corpus import webtext

# Make sure the necessary NLTK data is downloaded
nltk.download('webtext')

# Access words from two different web text sources
firefox_words = webtext.words(fileids='firefox.txt')
internet_words = webtext.words(fileids='wine.txt')

# Convert lists to sets to get unique words
firefox_types = set(firefox_words)
internet_types = set(internet_words)

# Print the number of words and types in each web text source
print(f"Number of words in firefox.txt: {len(firefox_words)}")
print(f"Number of unique words (types) in firefox.txt: {len(firefox_types)}")
print(f"Number of words in internet.txt: {len(internet_words)}")
print(f"Number of unique words (types) in internet.txt: {len(internet_types)}")
```

4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time

```python
import nltk
from nltk.corpus import state_union

# Make sure the necessary NLTK data is downloaded
nltk.download('state_union')

# Define the words to count
target_words = ['men', 'women', 'people']

# Initialize dictionaries to store counts
word_counts = {word: [] for word in target_words}
years = []

# Iterate over the State of the Union addresses
for fileid in state_union.fileids():
    # Extract the year from the file ID (e.g., '1790-Washington.txt' -> 1790)
    year = int(fileid.split('-')[0])
    years.append(year)
    
    # Tokenize the text
    words = state_union.words(fileid)
    
    # Count occurrences of each target word
    for word in target_words:
        count = words.count(word)
        word_counts[word].append(count)

print(f"{'Year':<10}", end='')
for word in target_words:
    print(f"{word:<10}", end='')
print()

for year, counts in zip(years, zip(word_counts['men'], word_counts['women'], word_counts['people'])):
    print(f"{year:<10}", end='')
    for count in counts:
        print(f"{count:<10}", end='')
    print()

# Optionally, you can visualize the trends using matplotlib (if installed)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for word in target_words:
    plt.plot(years, word_counts[word], marker='o', label=word)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Occurrences of Words in State of the Union Addresses Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

5. Investigate the holonym-meronym relations for some nouns. Remember that there are three kinds of holonym-meronym relation, so you need to use member_meronyms(), part_meronyms(), substance_meronyms(), member_holonyms(), part_holonyms(), and substance_holonyms().

```python
from nltk.corpus import wordnet as wn

# Ensure the necessary NLTK data is downloaded
import nltk
nltk.download('wordnet')

# Define a list of nouns to investigate
nouns = ['car', 'tree', 'house', 'team', 'table']

# Function to print holonym-meronym relations
def print_relations(word):
    synsets = wn.synsets(word, pos=wn.NOUN)
    
    if not synsets:
        print(f"No synsets found for {word}.")
        return
    
    for synset in synsets:
        print(f"\nSynset: {synset.name()}")
        
        # Meronyms
        member_meronyms = synset.member_meronyms()
        part_meronyms = synset.part_meronyms()
        substance_meronyms = synset.substance_meronyms()
        
        print("Member Meronyms:")
        for mm in member_meronyms:
            print(f"  - {mm.name()}: {mm.definition()}")
        
        print("Part Meronyms:")
        for pm in part_meronyms:
            print(f"  - {pm.name()}: {pm.definition()}")
        
        print("Substance Meronyms:")
        for sm in substance_meronyms:
            print(f"  - {sm.name()}: {sm.definition()}")
        
        # Holonyms
        member_holonyms = synset.member_holonyms()
        part_holonyms = synset.part_holonyms()
        substance_holonyms = synset.substance_holonyms()
        
        print("Member Holonyms:")
        for mh in member_holonyms:
            print(f"  - {mh.name()}: {mh.definition()}")
        
        print("Part Holonyms:")
        for ph in part_holonyms:
            print(f"  - {ph.name()}: {ph.definition()}")
        
        print("Substance Holonyms:")
        for sh in substance_holonyms:
            print(f"  - {sh.name()}: {sh.definition()}")

# Investigate each noun
for noun in nouns:
    print(f"\nInvestigating '{noun}':")
    print_relations(noun)
```

6.  In the discussion of comparative wordlists, we created an object called trans
late, which you could look up using words in both German and Italian in order to get corresponding words in English. What problem might arise with this approach? Can you suggest a way to avoid this problem?

```python
from nltk.corpus import wordnet as wn

def translate(word, source_lang, target_lang):
    # Get synsets for the word in the source language
    synsets = wn.synsets(word, lang=source_lang)
    
    if not synsets:
        return "No translation found."
    
    translations = []
    
    for synset in synsets:
        # Translate synset to the target language
        lemma_names = synset.lemma_names(target_lang)
        translations.extend(lemma_names)
    
    return set(translations)  # Return unique translations

# Example usage
source_word = 'Bank'
source_language = 'de'  # German
target_language = 'en'  # English

translations = translate(source_word, source_language, target_language)
print(f"Translations of '{source_word}' in English: {translations}")
```
