4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time

```python
import nltk
from nltk.corpus import state_union

# Make sure the necessary NLTK data is downloaded
nltk.download('state_union')

# Define the words to count
target_words = ['men', 'women', 'people']

# Initialize dictionaries to store counts
word_counts = {word: [] for word in target_words}
years = []

# Iterate over the State of the Union addresses
for fileid in state_union.fileids():
    # Extract the year from the file ID (e.g., '1790-Washington.txt' -> 1790)
    year = int(fileid.split('-')[0])
    years.append(year)
    
    # Tokenize the text
    words = state_union.words(fileid)
    
    # Count occurrences of each target word
    for word in target_words:
        count = words.count(word)
        word_counts[word].append(count)

print(f"{'Year':<10}", end='')
for word in target_words:
    print(f"{word:<10}", end='')
print()

for year, counts in zip(years, zip(word_counts['men'], word_counts['women'], word_counts['people'])):
    print(f"{year:<10}", end='')
    for count in counts:
        print(f"{count:<10}", end='')
    print()

# Optionally, you can visualize the trends using matplotlib (if installed)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for word in target_words:
    plt.plot(years, word_counts[word], marker='o', label=word)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Occurrences of Words in State of the Union Addresses Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

5. Investigate the holonym-meronym relations for some nouns. Remember that there are three kinds of holonym-meronym relation, so you need to use member_meronyms(), part_meronyms(), substance_meronyms(), member_holonyms(), part_holonyms(), and substance_holonyms().

```python
from nltk.corpus import wordnet as wn

# Ensure the necessary NLTK data is downloaded
import nltk
nltk.download('wordnet')

# Define a list of nouns to investigate
nouns = ['car', 'tree', 'house', 'team', 'table']

# Function to print holonym-meronym relations
def print_relations(word):
    synsets = wn.synsets(word, pos=wn.NOUN)
    
    if not synsets:
        print(f"No synsets found for {word}.")
        return
    
    for synset in synsets:
        print(f"\nSynset: {synset.name()}")
        
        # Meronyms
        member_meronyms = synset.member_meronyms()
        part_meronyms = synset.part_meronyms()
        substance_meronyms = synset.substance_meronyms()
        
        print("Member Meronyms:")
        for mm in member_meronyms:
            print(f"  - {mm.name()}: {mm.definition()}")
        
        print("Part Meronyms:")
        for pm in part_meronyms:
            print(f"  - {pm.name()}: {pm.definition()}")
        
        print("Substance Meronyms:")
        for sm in substance_meronyms:
            print(f"  - {sm.name()}: {sm.definition()}")
        
        # Holonyms
        member_holonyms = synset.member_holonyms()
        part_holonyms = synset.part_holonyms()
        substance_holonyms = synset.substance_holonyms()
        
        print("Member Holonyms:")
        for mh in member_holonyms:
            print(f"  - {mh.name()}: {mh.definition()}")
        
        print("Part Holonyms:")
        for ph in part_holonyms:
            print(f"  - {ph.name()}: {ph.definition()}")
        
        print("Substance Holonyms:")
        for sh in substance_holonyms:
            print(f"  - {sh.name()}: {sh.definition()}")

# Investigate each noun
for noun in nouns:
    print(f"\nInvestigating '{noun}':")
    print_relations(noun)
```

6.  In the discussion of comparative wordlists, we created an object called trans
late, which you could look up using words in both German and Italian in order to get corresponding words in English. What problem might arise with this approach? Can you suggest a way to avoid this problem?

```python
from nltk.corpus import wordnet as wn

def translate(word, source_lang, target_lang):
    # Get synsets for the word in the source language
    synsets = wn.synsets(word, lang=source_lang)
    
    if not synsets:
        return "No translation found."
    
    translations = []
    
    for synset in synsets:
        # Translate synset to the target language
        lemma_names = synset.lemma_names(target_lang)
        translations.extend(lemma_names)
    
    return set(translations)  # Return unique translations

# Example usage
source_word = 'Bank'
source_language = 'de'  # German
target_language = 'en'  # English

translations = translate(source_word, source_language, target_language)
print(f"Translations of '{source_word}' in English: {translations}")
```

```python
from nltk.corpus import wordnet as wn
from nltk.corpus.reader.wordnet import WordNetError

def translate(word, source_lang, target_lang):
    try:
        # Get synsets for the word in the source language
        synsets = wn.synsets(word, lang=source_lang)
        
        if not synsets:
            return "No translation found."
        
        translations = []
        
        for synset in synsets:
            # Translate synset to the target language
            lemma_names = synset.lemma_names(target_lang)
            translations.extend(lemma_names)
        
        return set(translations)  # Return unique translations
    
    except WordNetError as e:
        return str(e)

# Example usage
source_word = 'Bank'
source_language = 'de'  # German
target_language = 'en'  # English

translations = translate(source_word, source_language, target_language)
print(f"Translations of '{source_word}' in English: {translations}")
```

14.  Define a function supergloss(s) that takes a synset s as its argument and returns
a string consisting of the concatenation of the definition of s, and the definitions
of all the hypernyms and hyponyms of s.

```python
import nltk
from nltk.corpus import wordnet as wn

# Download WordNet if not already downloaded
nltk.download('wordnet')

def supergloss(synset):
    """
    Returns a string consisting of the definition of the synset
    and the definitions of all its hypernyms and hyponyms.
    """
    # Ensure the argument is a synset object
    if not isinstance(synset, wn.Synset):
        raise TypeError("Argument must be a WordNet Synset object")
    
    # Get the definition of the synset
    gloss = synset.definition()
    
    # Collect definitions of hypernyms
    hypernyms_definitions = [hypernym.definition() for hypernym in synset.hypernyms()]
    
    # Collect definitions of hyponyms
    hyponyms_definitions = [hyponym.definition() for hyponym in synset.hyponyms()]
    
    # Combine all definitions into a single string
    full_gloss = gloss + " " + " ".join(hypernyms_definitions) + " " + " ".join(hyponyms_definitions)
    
    return full_gloss

# Example usage
example_synset = wn.synset('dog.n.01')  # You can replace this with any other noun synset
print(supergloss(example_synset))
```

16.  Write a program to generate a table of lexical diversity scores (i.e., token/type

2.8 Exercises | 75(nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest

```python
import nltk
from nltk.corpus import brown
from nltk.probability import FreqDist

# Download Brown Corpus if not already downloaded
# nltk.download('brown')

# Get a list of genres from the Brown Corpus
genres = brown.categories()

# Initialize a dictionary to store token/type ratios
diversity_scores = {}

# Calculate the token/type ratio for each genre
for genre in genres:
    # Get words for the genre
    words = brown.words(categories=genre)
    
    # Calculate frequency distribution
    freq_dist = FreqDist(words)
    
    # Calculate number of tokens and types
    num_tokens = len(words)
    num_types = len(freq_dist)
    
    # Calculate token/type ratio
    token_type_ratio = num_tokens / num_types
    
    # Store the ratio in the dictionary
    diversity_scores[genre] = token_type_ratio

# Find the genre with the lowest diversity (highest token/type ratio)
lowest_diversity_genre = min(diversity_scores, key=diversity_scores.get)
lowest_diversity_score = diversity_scores[lowest_diversity_genre]

# Print the results
print(f'Lexical Diversity Scores (Token/Type Ratios):')
for genre, ratio in sorted(diversity_scores.items()):
    print(f'{genre:20}: {ratio:.2f}')

print(f'\nGenre with the lowest diversity (greatest number of tokens per type): {lowest_diversity_genre}')
print(f'Lexical Diversity Score for {lowest_diversity_genre}: {lowest_diversity_score:.2f}')
```

17.  Write a function that finds the 50 most frequently occurring words of a text that
are not stopwords.

```python
import nltk
from nltk.corpus import brown, stopwords
from nltk.probability import FreqDist

# Ensure required resources are downloaded
# nltk.download('brown')
# nltk.download('stopwords')

def find_most_freq_50_non_stopwords(text):
    # Create a frequency distribution of the words in the text
    fd = FreqDist(text)
    
    # Remove stopwords and non-alphabetic words
    filtered_fd = FreqDist(word for word in fd if word.isalpha() and word not in stopwords.words('english'))
    
    # Return the 50 most common words
    return filtered_fd.most_common(50)

# Load words from the 'news' category of the Brown Corpus
news_words = brown.words(categories='news')

# Find the 50 most frequently occurring non-stopwords
most_freq_50_non_stopwords = find_most_freq_50_non_stopwords(news_words)

# Print the results
print('50 most frequent non-stopwords:')
for word, freq in most_freq_50_non_stopwords:
    print(f'{word}: {freq}')
```

18.  Write a program to print the 50 most frequent bigrams (pairs of adjacent words)
of a text, omitting bigrams that contain stopwords.

```python
import nltk
from nltk.corpus import brown, stopwords
from nltk.util import bigrams
from nltk.probability import FreqDist

# Ensure required resources are downloaded
# nltk.download('brown')
# nltk.download('stopwords')

def find_most_freq_bigrams(text):
    # Generate bigrams from the text
    bigram_list = list(bigrams(text))
    
    # Define stopwords
    stop_words = set(stopwords.words('english'))
    
    # Filter out bigrams containing stopwords
    filtered_bigrams = [bigram for bigram in bigram_list if bigram[0] not in stop_words and bigram[1] not in stop_words]
    
    # Create a frequency distribution of the bigrams
    bigram_fd = FreqDist(filtered_bigrams)
    
    # Return the 50 most common bigrams
    return bigram_fd.most_common(50)

# Load words from the 'news' category of the Brown Corpus
news_words = brown.words(categories='news')

# Find the 50 most frequent bigrams that do not contain stopwords
most_freq_bigrams = find_most_freq_bigrams(news_words)

# Print the results
print('50 most frequent bigrams (excluding stopwords):')
for bigram, freq in most_freq_bigrams:
    print(f'{bigram}: {freq}')
```

```python
import nltk
from nltk.corpus import brown
from nltk.probability import FreqDist

# Ensure that the Brown Corpus is downloaded
# nltk.download('brown')

def word_freq(word, section):
    """
    Compute the frequency of a word in a specified section of the Brown Corpus.
    
    Parameters:
    word (str): The word to search for.
    section (str): The section of the Brown Corpus (e.g., 'news', 'fiction').
    
    Returns:
    int: The frequency of the word in the specified section.
    """
    # Get the list of words in the specified section
    words = brown.words(categories=section)
    
    # Create a frequency distribution of words
    freq_dist = FreqDist(words)
    
    # Return the frequency of the given word
    return freq_dist[word]

# Example usage
word = 'data'
section = 'news'
print(f"The word '{word}' appears {word_freq(word, section)} times in the '{section}' section.")
```

```python
import nltk
from nltk.corpus import cmudict
from nltk.tokenize import word_tokenize

# Ensure that the CMU Pronouncing Dictionary and tokenizer are downloaded
# nltk.download('cmudict')
# nltk.download('punkt')

# Load the CMU Pronouncing Dictionary
d = cmudict.dict()

def count_syllables(word):
    """
    Count the number of syllables in a word using the CMU Pronouncing Dictionary.
    
    Parameters:
    word (str): The word to count syllables for.
    
    Returns:
    int: The number of syllables in the word. Returns 0 if the word is not found.
    """
    word = word.lower()
    if word in d:
        # Get the list of pronunciations for the word
        pronunciations = d[word]
        # Count the number of syllables in each pronunciation
        syllable_counts = [
            max([
                sum([1 for phoneme in pronunciation if phoneme[-1].isdigit()])
                for pronunciation in pronunciations
            ])
        ]
        # Return the maximum syllable count among all pronunciations
        return syllable_counts[0]
    else:
        return 0

def text_syllable_count(text):
    """
    Count the total number of syllables in a text.
    
    Parameters:
    text (str): The text to analyze.
    
    Returns:
    int: The total number of syllables in the text.
    """
    # Tokenize the text into words
    words = word_tokenize(text)
    # Count the syllables for each word and sum them up
    return sum(count_syllables(word) for word in words)

# Example usage
text = "Hello world! This is a test sentence to count syllables."
print(f"The text contains {text_syllable_count(text)} syllables.")
```

22.  Define a function hedge(text) that processes a text and produces a new version
with the word 'like' between every third word.

```python
def hedge(text):
    # test = 'this is a test sentence to insert like after every third word'.split()
    ids = [index-1 for index in list(range(3,len(text)+1,3))]
    for id in ids:
        text.insert(id,'like')
    return text

test_text=hedge('this is a test sentence to insert like after every third word'.split())
print(test_text)
```

25.  Define a function find_language() that takes a string as its argument and returns
a list of languages that have that string as a word. Use the udhr corpus and limit
your searches to files in the Latin-1 encoding

```python
import nltk
from nltk.corpus import udhr

# Ensure the UDHR corpus is downloaded
# nltk.download('udhr')

def find_language(search_string):
    """
    Find languages in the UDHR corpus that contain the given string as a word.
    
    Parameters:
    search_string (str): The string to search for.
    
    Returns:
    list: A list of languages that contain the given string.
    """
    # List to hold languages where the search_string is found
    languages_with_string = []
    
    # Iterate over files in the UDHR corpus
    for fileid in udhr.fileids():
        # Check if the file is in Latin-1 encoding
        if fileid.endswith('.latin1'):
            # Read the content of the file
            text = udhr.raw(fileid)
            # Split the text into words and check for the search_string
            words = text.split()
            if search_string in words:
                # If found, add the language to the list
                languages_with_string.append(fileid.split('/')[0])
    
    return languages_with_string

# Example usage
search_string = "human"
print(f"Languages containing the word '{search_string}': {find_language(search_string)}")
```

```python
import nltk
import re
from nltk.corpus import udhr

# Ensure the UDHR corpus is downloaded
# nltk.download('udhr')

# List of Latin-1 encoded files
latin1_files = [f for f in udhr.fileids() if re.search(r'Latin1', f)]

def find_language(word, latin1_files):
    """
    Count the occurrences of a word in Latin-1 encoded UDHR files.
    
    Parameters:
    word (str): The word to search for.
    latin1_files (list): List of Latin-1 encoded files in the UDHR corpus.
    
    Returns:
    int: The total number of occurrences of the word.
    """
    return sum([1 for file in latin1_files for w in udhr.words(file) if word in w])

# Example usage
word_count_latin1 = find_language("human", latin1_files)
print(f"Occurrences of 'human': {word_count_latin1}")

word_count_latin1 = find_language("rights", latin1_files)
print(f"Occurrences of 'rights': {word_count_latin1}")
```

26.  What is the branching factor of the noun hypernym hierarchy? I.e., for every
noun synset that has hyponymsor children in the hypernym hierarchyhow
many do they have on average? You can get all noun synsets using wn.all_syn
sets('n')

```python
import nltk
from nltk.corpus import wordnet as wn

# Ensure that the WordNet data is downloaded
# nltk.download('wordnet')

def branching_factor():
    """
    Calculate the branching factor of the noun hypernym hierarchy in WordNet.
    
    Returns:
    float: The average number of hyponyms per noun synset.
    """
    # Get all noun synsets
    noun_synsets = list(wn.all_synsets('n'))
    
    # Calculate the number of hyponyms for each noun synset
    total_hyponyms = 0
    num_synsets_with_hyponyms = 0
    
    for synset in noun_synsets:
        hyponyms = synset.hyponyms()
        if hyponyms:
            total_hyponyms += len(hyponyms)
            num_synsets_with_hyponyms += 1
    
    # Calculate the average number of hyponyms per noun synset
    if num_synsets_with_hyponyms > 0:
        avg_hyponyms = total_hyponyms / num_synsets_with_hyponyms
    else:
        avg_hyponyms = 0
    
    return avg_hyponyms

# Example usage
print(f"The average number of hyponyms per noun synset is {branching_factor():.2f}")
```
